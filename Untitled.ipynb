```python
import subprocess
import graph_tool.all as gt
import numpy as np
import tqdm
import random as rnd
import matplotlib.pyplot as plt
import pylab as pl
import pandas as pd
import numpy as np
from collections import Counter
import pickle
import time
from joblib import Parallel, delayed
import pylab as pl
from IPython import display

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 

```

```python
spexs_df = pd.read_csv('data/specs_out.txt',sep='\t')

patterns_df = spexs_df.loc[np.array(list(map(len,spexs_df['Pat?()'])))  == 6].sort('Pat?()')

df = pd.read_csv('data/protobios_generated_data/peptides_s50_r50_m0.txt',sep='\t', header=None)[1]
#df = pd.read_csv('data/protobios_generated_data/peptides_s50_r0_m0.txt',sep='\t', header=None)[1]

PEPTIDE_LENGTH = 4
peptides = list()
for protein in df:
    for i in range(12 - PEPTIDE_LENGTH + 1):
        peptides.append(protein[i:i+PEPTIDE_LENGTH])

print(len(peptides))
```

```python
testLength = len(peptides)
#testLength = 200
```

```python




g = gt.Graph()

e_weight = g.new_edge_property("int")
g.edge_properties['weights'] = e_weight
e_colored = g.new_edge_property("int")
g.edge_properties['colors'] = e_colored
v_letter = g.new_vertex_property('string')
g.vertex_properties['letters'] = v_letter

root = g.add_vertex(1)
for i in tqdm.tqdm(range(testLength)):
    peptide = peptides[i]
    current = root
    for letter in peptide:
        arr = np.array([v_letter[x] == letter for x in current.out_neighbours()])
        if any(arr):
            target_index = arr.argmax()
            target = list(current.out_neighbours())[target_index]
        else:
            target = g.add_vertex()
            edge = g.add_edge(current,target)
            e_weight[edge] = 0
            e_colored[edge] = 0
            v_letter[target] = letter
        edge = g.edge(current,target)
        e_weight[edge] += 1
        current = target
```

```python
patterns = patterns_df['Pat?()']

#print(len(patterns))

def ifBetterThanRandom(amountOfPatterns,weight, depth):
    threshold = amountOfPatterns*(1/20)**depth
    #print(depth)
    #print(weight,threshold)
    if weight > threshold:
        return True
    else:
        return False

for pattern in patterns:
    vertices = set()
    nextVertices = set()
    nextVertices.add(root)
    depth = 0
    for letter in pattern:
        vertices = nextVertices.copy()
        nextVertices = set()
        depth += 1
        for current in vertices:
            if letter == '.':
                #print('Found points')
                for neighbor in list(current.out_neighbours()):
                    edge = g.edge(current,neighbor)
                    weight = e_weight[edge]
                    #print(weight)
                    if ifBetterThanRandom(testLength, weight, depth):
                        e_colored[edge] = 1
                        nextVertices.add(neighbor)
                    #else:
                    #    print("NOT")
            else:
                try:
                    neighbor = [x for x in current.out_neighbours() if v_letter[x] == letter][0]
                except IndexError:
                    continue
                #print("NEW")
                #print(letter)
                #print(v_letter[neighbor])
                edge = g.edge(current,neighbor)
                weight = e_weight[edge]
                if ifBetterThanRandom(testLength, weight, depth):
                    e_colored[edge] = 1
                    nextVertices.add(neighbor)
                #else:
                #    print("NOT2")
```

```python
#
# Global variables
# Setup optimal string and GA input variables.
#





def grep(filename, mismatch,regex):
    regex = ['[{}]'.format(x) if len(x)>1 else x for x in regex]
    process = subprocess.Popen(['tre-agrep', '-E', mismatch, '-e={}'.format(regex), filename],
                               stdout=subprocess.PIPE, universal_newlines=True)
    stdout,stderr = process.communicate()
    matches = sum(map(int,[x.split('\t')[0] for x in stdout.split('\n')[:-1]]))
    
    return matches


def possible_strings(individual):
    n = 0
    for group in individual:
        if len(group) == 1:
            if group == '.':
                n *= 20
            else:
                n += 1
        else:
            n *= len(group)
    return n
            

# GA functions
# These make up the bulk of the actual GA algorithm.

def get_fitness(individual):
    individual = ['[{}]'.format(x) if len(x)>1 else x for x in individual]
    matches = grep(filename,MISMATCHES,''.join(individual))
    specificity = possible_strings(individual)
    
    return matches/specificity

def remove_letter(group):
    """
    Helper function for mutation.
    Removes one letter from a group.
    If letter is the last one in group,
    returns empty string
    """
    if group == '.':
        group = ''.join(np.random.choice(list(ALPHABET),19))
    elif len(group) == 1:
        return ''
    else:
        group = ''.join(np.random.choice(list(group),len(group)-1))
    return group

def add_letter_rec(individual,i,j,vertex):
    """
    Helper function for add_letter.
    Finds all compatible nodes for an individual
    on requested level
    """
    if i == j:
        return [x.target() for x in vertex.out_edges() if e_colored[x] == 1]
    result=[]
    for v in [x.target() for x in vertex.out_edges() if e_colored[x] == 1]:
        result.extend(add_letter_rec(individual, i, j+1, v))
    return result


def add_letter(individual,depth):
    """
    Helper function for mutate.
    Gets all compatible nodes on the required level
    and returns a letter chosen with weighed random
    function where weight is the probability of
    the letter considering the preceeding regex
    """
    candidates = add_letter_rec(individual,depth,0,root)
    lcounts = {}
    for c in candidates:
        if v_letter[c] in lcounts:
            lcounts[v_letter[c]] += e_weight[list(c.in_edges())[0]]
        else:
            lcounts[v_letter[c]] = e_weight[list(c.in_edges())[0]]
    letters = []
    weigths = np.array([])
    for k,v in lcounts.items():
        letters.append(k)
        np.append(weigths,v)
    try:
        return np.random.choice(letters,1,weigths/weigths.sum())[0]
    except ValueError:
        return '' 

         
    
    
def mutate(individual):
    """
    Function that performs mutation on individuals
    with the given mutation_chance probability
    Chooses randomly from three different mechanisms
    """
    m_individual = []
    mutation_chance = 0.01
    for i in range(len(individual)):
        c = individual[i]
        if np.random.randint(0,100)/100 < mutation_chance:
            mut_choice = np.random.random()
            if mut_choice<0.33 and len(c) > 1: # Remove letter from group
                c = remove_letter(c)
            if 0.33<mut_choice<0.66: # Add letter to group
                if c == '.':
                    m_individual.append(c)
                    continue
                n_letter = add_letter(individual,i)
                c += n_letter
            #else:
            #    c = switch_letter(c)
        if c == '':
            continue
        m_individual.append(c)
    return m_individual

def crossover(individual1,individual2):
    """
    Crossover function.
    Switches random groups
    between two individuals
    """
    idx = np.random.choice(range(len(individual1)),0.3*len(individual1),replace=False)
    old_individual1 = individual1[:]
    try:
        for i in idx:
            individual1[i] = individual2[i]
            individual2[i] = individual1[i]
    except Exception as e:
        print(e)
        print('Error with individuals {} and {}'.format(individual1,individual2))
    return [individual1,individual2]

candidates = pd.read_csv('data/specs_out.txt',sep='\t')['Pat?()'].values
    
```

```python
individuals = [list(x) for x in candidates]

CANDIDATES = 200
GENERATIONS = 10000
MAX_LENGTH = 20**8
FILE_LENGTH = df.shape[0]
MISMATCHES = '1'
ALPHABET = 'GPAVLIMCFYWHKRQNEDST'
filename='data/protobios_generated_data/peptides_s50_r0_m0.txt'
scores = []

with Parallel(n_jobs=3) as parallel:
    for iteration in range(11):
        matches = parallel(delayed(grep)(filename,'1',individual) for individual in individuals)    
        nrOfStrings = parallel(delayed(possible_strings)(individual) for individual in individuals)
        #fitnesses = np.array(matches)/np.log10(np.array(nrOfStrings))*np.array(list(map(len,individuals)))
        #fitnesses = (fitnesses - fitnesses.min())/(fitnesses.max()-fitnesses.min())
        fitnesses = (np.array(matches)/FILE_LENGTH) + (1-np.array(nrOfStrings)/MAX_LENGTH)
        
        weigths = fitnesses/fitnesses.sum()
        new_population = []
        while len(new_population)<CANDIDATES:
            i1,i2 = np.random.choice(list(range(len(individuals))),2,replace=False,p=weigths)
            c1=individuals[i1]
            c2=individuals[i2]
            if len(c1) == len(c2):
                new_population.extend(crossover(c1, c2))
                new_population.extend([c1,c2])
            
        individuals = [mutate(x) for x in new_population]

        current_score = np.mean(fitnesses)
        scores.append(current_score)
        if iteration % 10 == 0:
            pl.plot(list(range(len(scores))),scores)
            display.clear_output(wait=True)
            display.display(pl.gcf())
            time.sleep(1.0)
```

```python
def get_edges(individual,vertex,i):
    if i == len(individual):
        return
    edges = []
    group = individual[i]
    if group == '.':
        edges.extend(list(vertex.out_edges()))
        for vertex in vertex.out_neighbours():
            d=get_edges(individual,vertex,i+1)
            if d != None:
                edges.extend(d)
    else:
        validEdges = []
        for letter in group:
            validEdges.extend([x for x in vertex.out_edges() if v_letter[x.target()] == letter])
        for edge in validEdges:
            d = get_edges(individual,edge.target(),i+1)
            if d != None:
                edges.extend(d)
        edges.extend(validEdges)
            
    return set(edges)


s1 = get_edges(individuals[3],root,0)
s2 = get_edges(individuals[4],root,0)

print(s1.intersection(s2))

def find_overlap(c1,c2):
    return len(get_edges(c1,root,0).intersection(get_edges(c2,root,0)))
```
