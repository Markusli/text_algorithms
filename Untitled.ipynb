```python
import subprocess
import graph_tool.all as gt
import numpy as np
import tqdm
import random as rnd
import matplotlib.pyplot as plt
import pylab as pl
import pandas as pd
import numpy as np
from collections import Counter
import pickle
import time
import copy
from joblib import Parallel, delayed
from IPython import display
%matplotlib inline
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 

#CONSTANTS--------------------------------
filename='data/protobios_generated_data/peptides_s50_r50_m0.txt'
PEPTIDE_LENGTH = 6
spexs_df = pd.read_csv('data/specs_out.txt',sep='\t')
patterns_df = spexs_df.loc[np.array(list(map(len,spexs_df['Pat?()'])))  == 6].sort('Pat?()')
graph_filename = 'full_graph_pep_len_%s.p' % PEPTIDE_LENGTH
ALPHABET = 'GPAVLIMCFYWHKRQNEDST'
#CONSTANTS--------------------------------

#from IPython.core.display import display, HTML
#display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
#READ IN PEPTIDES--------------------------------
df = pd.read_csv(filename,sep='\t', header=None)[1]

peptides = list()
for protein in df:
    for i in range(12 - PEPTIDE_LENGTH + 1):
        peptides.append(protein[i:i+PEPTIDE_LENGTH])
peptidesToAdd = len(peptides)
```

```python
#CONSTRUCT GRAPH--------------------------------
g = gt.Graph()

e_weight = g.new_edge_property("int")
g.edge_properties['weights'] = e_weight
e_colored = g.new_edge_property("int")
g.edge_properties['colors'] = e_colored
v_letter = g.new_vertex_property('string')
g.vertex_properties['letters'] = v_letter
v_depth = g.new_vertex_property('int')
g.vertex_properties['depths'] = v_depth

root = g.add_vertex(1)
v_depth[root] = 0
for i in tqdm.tqdm(range(peptidesToAdd)):
    peptide = peptides[i]
    current = root
    current_depth = 0
    for letter in peptide:
        current_depth += 1
        arr = np.array([v_letter[x] == letter for x in current.out_neighbours()])
        if any(arr):
            target_index = arr.argmax()
            target = list(current.out_neighbours())[target_index]
        else:
            target = g.add_vertex()
            edge = g.add_edge(current,target)
            e_weight[edge] = 0
            e_colored[edge] = 0
            v_letter[target] = letter
            v_depth[target] = current_depth
        edge = g.edge(current,target)
        e_weight[edge] += 1
        current = target
pickle.dump(g, open(graph_filename, 'wb'))
```

```python
#LOAD GRAPH--------------------------------
g = pickle.load(open(graph_filename, 'rb'))
root = [x for x in list(g.vertices()) if x.in_degree() == 0][0]
e_weight = g.edge_properties['weights']
e_colored = g.edge_properties['colors']
v_letter = g.vertex_properties['letters']
v_depth = g.vertex_properties['depths']
```

```python
#LETTER WEIGHTS AT GRAPH DEPTHS--------------------------------
letterWeights = {i: {c: 0 for c in ALPHABET} for i in range(1, PEPTIDE_LENGTH+1)}
for v in g.vertices():
    for neighbour in list(v.out_neighbours()):
        edge = g.edge(v,neighbour)
        weight = e_weight[edge]
        letter = v_letter[neighbour]
        depth = v_depth[neighbour]
        letterWeights[depth][letter] += weight
#LETTER PROBABILITIES AT GRAPH DEPTHS--------------------------------
letterProbabilities = copy.deepcopy(letterWeights)
for depth, letters in letterWeights.items():
    print(depth, letters)
    lettersTotal = sum(letters.values())
    for letter, weight in letters.items():
        letterProbabilities[depth][letter] = weight/lettersTotal
```

```python
#UTILITY FUNCTIONS--------------------------------

def file_len(fname):
    with open(fname) as f:
        for i, l in enumerate(f):
            pass
    return i + 1

def grepWithLineNumbers(filename, mismatch, regex):
    regex = ['[{}]'.format(x) if len(x)>1 else x for x in regex]
    regex = ''.join(regex)
    process = subprocess.Popen(['agrep', '-n','-'+mismatch, regex, filename],
                               stdout=subprocess.PIPE, universal_newlines=True)
    stdout,stderr = process.communicate()
    lines = map(int,[x.split('\t')[0].split(':')[0] for x in stdout.split('\n')[:-1]])
    lines = []
    matches = 0
    for line in stdout.split('\n')[:-1]:
        splitLine = line.split('\t')[0].split(':')
        lines.append(int(splitLine[0]))
        try:
            matches += int(splitLine[1])
        except ValueError:
            matches+=1
    return matches, set(lines)

def possible_strings(individual):
    n = 0
    for group in individual:
        if len(group) == 1:
            if group == '.':
                n *= 20
            else:
                n += 1
        else:
            n *= len(group)
    return n

def separate_weighted_dict(dictionary):
    keysAndWeights = list(zip(*dictionary.items()))
    keys = keysAndWeights[0]
    weights = keysAndWeights[1]
    return keys, weights

# GA functions
# These make up the bulk of the actual GA algorithm.

def remove_letter(group,depth):
    """
    Helper function for mutation.
    Removes one letter from a group.
    If letter is the last one in group,
    returns empty string
    """
    if group == '.':

        letters, weights = separate_weighted_dict(letterProbabilities[depth])
        group = ''.join(np.random.choice(letters,5,replace=False,p=weights))
    elif len(group) == 1:
        return group
    else:
        group = list(group)
        del group[get_least_probable_index(group, depth)]
        group = ''.join(group)
    return group

def get_least_probable_index(group, depth):
    groupIndexProbabilities = {index: 1-letterProbabilities[depth][letter] for index, letter in enumerate(group)}
    probabilitySum = sum(groupIndexProbabilities.values())
    for index, probability in groupIndexProbabilities.items():
        groupIndexProbabilities[index] = probability/probabilitySum
    indexes, weights = separate_weighted_dict(groupIndexProbabilities)
    return np.random.choice(indexes, 1, p=weights)[0]

def get_letter_at_depth(depth):
    letters, weights = separate_weighted_dict(letterProbabilities[depth])
    return np.random.choice(letters,1,replace=False,p=weights)[0]

def change_letter(group,depth):
    group = list(group)
    n_letter = get_letter_at_depth(depth)
    while n_letter in group:
        n_letter = get_letter_at_depth(depth)
    group[get_least_probable_index(group, depth)] = n_letter
    group = ''.join(group)
    return group

def mutate(individual):
    """
    Function that performs mutation on individuals
    with the given mutation_chance probability
    Chooses randomly from three different mechanisms
    """
    m_individual = []
    mutation_chance = 1/len(individual)
    for depth in range(1, len(individual) + 1):
        c = individual[depth-1]
        if np.random.randint(0,100)/100 < mutation_chance:
            mut_choice = np.random.random()
            if mut_choice < 0.33: # Remove letter from group
                c = remove_letter(c,depth)
            elif 0.33 <= mut_choice < 0.66: # Add letter to group
                if c == '.':
                    m_individual.append(c)
                    continue
                n_letter = get_letter_at_depth(depth)
                c = ''.join(set(c).union(set(n_letter)))
            else:#CHANGING LETTER
                if c == '.':
                    c = remove_letter(c,depth)
                else:
                    c = change_letter(c, depth)
        m_individual.append(c)
    return m_individual

def crossover(individual1,individual2):
    """
    Crossover function.
    Switches random groups
    between two individuals
    """
    idx = np.random.choice(range(len(individual1)),0.3*len(individual1),replace=False)
    old_individual1 = individual1[:]
    try:
        for i in idx:
            individual1[i] = individual2[i]
            individual2[i] = old_individual1[i]
    except Exception as e:
        print(e)
        print('Error with individuals {} and {}'.format(individual1,individual2))
    return [individual1,individual2]

def crossover_at_index(individual1,individual2):
    idx = np.random.choice(range(len(individual1)-1),1,replace=False)
    old_individual1 = individual1[:]
    individual1[0:idx] = individual2[0:idx]
    individual2[0:idx] = old_individual1[0:idx]
    return [individual1,individual2]

candidates = pd.read_csv('data/specs_out.txt',sep='\t')['Pat?()'].values
    
```

```python
individuals = [list(x) for x in candidates if len(x) == PEPTIDE_LENGTH]
initial_individuals = np.array(individuals[:],dtype='<U6')
#initial_individuals = individuals[:]

POPULATION_SIZE = 200
GENERATIONS = 250
FILE_LENGTH = df.shape[0]
MISMATCHES = '1'
scores = []
keepBest = 0.15
keepOld = 0.40
howManyToMutate = 0.025
similarityThreshold = 0.95
fileLength = file_len(filename)
fileLengthThreshold = 0.025
matchesThreshold = fileLength*fileLengthThreshold

with Parallel(n_jobs=3) as parallel:
    individualsKept = np.ones(len(individuals))
    newIndividuals = np.ones(len(individuals))
    matches = np.zeros(len(individuals))
    nrOfStrings = np.zeros(len(individuals))
    individuals = np.array(individuals,dtype='<U6')
    rowArrays = [0] * len(individuals)
    similarityMatrix = np.zeros([len(individuals),len(individuals)])
    for iteration in range(GENERATIONS):
        #MERGE OF SIMILAR
        for i in range(similarityMatrix.shape[0]):
            similarityMatrix[i,i] = 0
            if matches[i] >= matchesThreshold:
                similarityMatrix[i,:] = 0
                similarityMatrix[:,i] = 0
        while np.any(similarityMatrix > similarityThreshold):
            row,column = np.unravel_index(similarityMatrix.argmax(), similarityMatrix.shape)
            smallerRegexp = individuals[row]
            biggerRegexp = individuals[column]
            combinedRegexp = []
            for i in range(len(smallerRegexp)):
                if smallerRegexp[i] == biggerRegexp[i]:
                    combinedRegexp.append(smallerRegexp[i])
                    continue
                if biggerRegexp[i] == '.':
                    combinedRegexp.append('.')
                    continue
                if smallerRegexp[i] == '.':
                    combinedRegexp.append(biggerRegexp[i])
                    continue        
                combinedRegexp.append(''.join(set(smallerRegexp[i]).union(set(biggerRegexp[i]))))
            individuals[column] = combinedRegexp
            newIndividuals[column] = 1
            newIndividuals[row] = 0
            individualsKept[row] = 0
            similarityMatrix[row,:] = 0
            similarityMatrix[:,row] = 0
            similarityMatrix[column,:] = 0
            similarityMatrix[:,column] = 0
        #END OF MERGE OF SIMILAR

        #MUTATION
        indicesToMutate = np.random.choice(range(len(individuals)), howManyToMutate*len(individuals))
        individuals[indicesToMutate] = np.array([mutate(x) for x in individuals[indicesToMutate]])
        newIndividuals[indicesToMutate] = 1
        #END OF MUTATION
        
        #CALCULATION OF NEW MEASURES
        grepResults = parallel(delayed(grepWithLineNumbers)(filename,MISMATCHES,individual) for individual in individuals[newIndividuals ==1])
        nrOfStrings[newIndividuals == 1] = parallel(delayed(possible_strings)(individual) for individual in individuals[newIndividuals ==1])
        matchesResult = []
        grepRows = []
        for grepResult in grepResults:
            matchesResult.append(grepResult[0])
            grepRows.append(grepResult[1])
        matches[newIndividuals == 1] = matchesResult
        
        for idx, value in zip(np.argsort(newIndividuals, kind='mergesort')[::-1][:np.sum(newIndividuals)][::-1],grepRows):
            rowArrays[idx] = value
        
        for idx in np.where(individualsKept==1)[0]:
            for newIdx in np.where(newIndividuals==1)[0]:
                similarityMatrix[idx,newIdx] = len(rowArrays[idx].intersection(rowArrays[newIdx]))/max([1, len(rowArrays[idx])])
        for idx in np.where(newIndividuals==1)[0]:
            for newIdx in np.where((newIndividuals==1) | (individualsKept==1))[0]:
                similarityMatrix[idx,newIdx] = len(rowArrays[idx].intersection(rowArrays[newIdx]))/max([1, len(rowArrays[idx])])
        #END OF CALCULATION OF NEW MEASURES
        
        #BEGINNING (OF THE END) OF DELETION OF UNNECESSARY!!!!!!
        matches = matches[(newIndividuals == 1) | (individualsKept == 1)]
        nrOfStrings = nrOfStrings[(newIndividuals == 1) | (individualsKept == 1)]
        individuals = individuals[(newIndividuals == 1) | (individualsKept == 1)]
        indicesToDelete = np.where((newIndividuals==0) & (individualsKept==0))[0][::-1]
        for idx in indicesToDelete:
            del rowArrays[idx]
        similarityMatrix = np.delete(similarityMatrix, indicesToDelete, axis=0)
        similarityMatrix = np.delete(similarityMatrix, indicesToDelete, axis=1)
        #END OF DELETION OF UNNECESSARY!!!!!!
        
                
        #CALCULATION OF FITNESS
        matchesScore = (np.array(matches)/FILE_LENGTH)
        matchesScore = (matchesScore-matchesScore.min())/(matchesScore.max()-matchesScore.min())#DIVISION BY ZERO
        complexityScore = 1-(np.array(nrOfStrings)/(20**np.array(list(map(len,individuals)))))
        complexityScore = (complexityScore-complexityScore.min())/(complexityScore.max()-complexityScore.min())
        
        similarityOverThreshold = np.zeros((len(individuals), len(individuals)))
        similarityOverThreshold[np.where(similarityMatrix > similarityThreshold)] = similarityMatrix[np.where(similarityMatrix > similarityThreshold)]
        similarityScore = np.sum(similarityOverThreshold, axis=0)
        similarityScore = (similarityScore-similarityScore.min())/(similarityScore.max()-similarityScore.min())
        similarityScore = 1 - similarityScore
        fitnesses =  matchesScore + complexityScore + similarityScore
        
        weigths = fitnesses
        weigths = weigths/weigths.sum()
        #END OF CALCULATION OF FITNESS
        
        #DONORS INTO POPULATION
        
        populationSizeToAdd = POPULATION_SIZE - len(individuals)
        donor_indexes = np.random.choice(range(len(initial_individuals)), populationSizeToAdd,replace=False)
        donors_from_initial = initial_individuals[donor_indexes]

        individuals = np.insert(individuals, (len(individuals),) * populationSizeToAdd, donors_from_initial, axis=0)
        similarityMatrix = np.insert(similarityMatrix, (similarityMatrix.shape[1],) * populationSizeToAdd, 0, axis=1)
        similarityMatrix = np.insert(similarityMatrix, (similarityMatrix.shape[0],) * populationSizeToAdd, 0, axis=0)
        matches = np.insert(matches, (len(matches),) * populationSizeToAdd, 0, axis=0)
        nrOfStrings = np.insert(nrOfStrings, (len(nrOfStrings),) * populationSizeToAdd, 0, axis=0)
        rowArrays = list(np.insert(rowArrays, (len(rowArrays),) * populationSizeToAdd, set(), axis=0))
        #individualsKept = np.zeros(len(individuals))
        individualsKept = np.ones(len(individuals))
        newIndividuals = np.zeros(len(individuals))
        newIndividuals[:-populationSizeToAdd] = 1
        #END OF DONORS INTO POPULATION
        
        """
        
        kept = max([keepBest*POPULATION_SIZE, keepBest*len(individuals)])
        indicesToKeep = np.argsort(weigths)[::-1][:kept]
        individualsKept[indicesToKeep] = 1
        indicePairs = []
        
        for index in indicesToKeep:
            i1 = index
            i2 = np.random.choice(list(range(len(individuals))),1,replace=False,p=weigths)[0]
            indicePairs.append((i1,i2))
            individualsKept[[i1,i2]] = 1

        while np.sum(individualsKept) < max([keepOld*POPULATION_SIZE, keepOld*len(individuals)]):
            i1,i2 = np.random.choice(list(range(len(individuals))),2,replace=False,p=weigths)
            indicePairs.append((i1,i2))
            individualsKept[[i1,i2]] = 1
            #if np.sum(individualsKept) > max([0.45*POPULATION_SIZE, 0.45*len(individuals)]):
            #    individualsKept[[i1,i2]] = 0 
            #    indicePairs = indicePairs[:-1]

        for indicePair in indicePairs:
            c1=individuals[indicePair[0]]
            c2=individuals[indicePair[1]]
            indices = np.argsort(individualsKept)[:2]
            individualsKept[indices] = 1
            individuals[indices] = crossover_at_index(c1, c2)
            newIndividuals[indices] = 1
        """
        
        current_score = np.mean(fitnesses)#calculated based on old population
        scores.append(current_score)
        #if iteration % 10 == 0:
        pl.plot(list(range(len(scores))),scores, color='green')
        display.clear_output(wait=True)
        display.display(pl.gcf())
        print(len(individuals))
        print(populationSizeToAdd)
        #print(individuals)
        #print(individuals[indicesToMutate])
        time.sleep(1.0)
```
