```python
import subprocess
import graph_tool.all as gt
import numpy as np
import tqdm
import random as rnd
import matplotlib.pyplot as plt
import pylab as pl
import pandas as pd
import numpy as np
from collections import Counter
import pickle
import time
import copy
from joblib import Parallel, delayed
from IPython import display
%matplotlib inline
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning) 

#CONSTANTS--------------------------------
filename='data/protobios_generated_data/peptides_s50_r50_m0.txt'
PEPTIDE_LENGTH = 6
spexs_df = pd.read_csv('data/specs_out.txt',sep='\t')
patterns_df = spexs_df.loc[np.array(list(map(len,spexs_df['Pat?()'])))  == 6].sort('Pat?()')
graph_filename = 'full_graph_pep_len_%s.p' % PEPTIDE_LENGTH
ALPHABET = 'GPAVLIMCFYWHKRQNEDST'
#CONSTANTS--------------------------------

#from IPython.core.display import display, HTML
#display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
#READ IN PEPTIDES--------------------------------
df = pd.read_csv(filename,sep='\t', header=None)[1]

peptides = list()
for protein in df:
    for i in range(12 - PEPTIDE_LENGTH + 1):
        peptides.append(protein[i:i+PEPTIDE_LENGTH])
peptidesToAdd = len(peptides)
```

```python
#CONSTRUCT GRAPH--------------------------------
g = gt.Graph()

e_weight = g.new_edge_property("int")
g.edge_properties['weights'] = e_weight
e_colored = g.new_edge_property("int")
g.edge_properties['colors'] = e_colored
v_letter = g.new_vertex_property('string')
g.vertex_properties['letters'] = v_letter
v_depth = g.new_vertex_property('int')
g.vertex_properties['depths'] = v_depth

root = g.add_vertex(1)
v_depth[root] = 0
for i in tqdm.tqdm(range(peptidesToAdd)):
    peptide = peptides[i]
    current = root
    current_depth = 0
    for letter in peptide:
        current_depth += 1
        arr = np.array([v_letter[x] == letter for x in current.out_neighbours()])
        if any(arr):
            target_index = arr.argmax()
            target = list(current.out_neighbours())[target_index]
        else:
            target = g.add_vertex()
            edge = g.add_edge(current,target)
            e_weight[edge] = 0
            e_colored[edge] = 0
            v_letter[target] = letter
            v_depth[target] = current_depth
        edge = g.edge(current,target)
        e_weight[edge] += 1
        current = target
pickle.dump(g, open(graph_filename, 'wb'))
```

```python
#LETTER WEIGHTS AT GRAPH DEPTHS--------------------------------
letterWeights = {i: {c: 0 for c in ALPHABET} for i in range(1, PEPTIDE_LENGTH+1)}
for v in g.vertices():
    for neighbour in list(v.out_neighbours()):
        edge = g.edge(v,neighbour)
        weight = e_weight[edge]
        letter = v_letter[neighbour]
        depth = v_depth[neighbour]
        letterWeights[depth][letter] += weight
#LETTER PROBABILITIES AT GRAPH DEPTHS--------------------------------
letterProbabilities = copy.deepcopy(letterWeights)
for depth, letters in letterWeights.items():
    print(depth, letters)
    lettersTotal = sum(letters.values())
    for letter, weight in letters.items():
        letterProbabilities[depth][letter] = weight/lettersTotal
```

```python
#UTILITY FUNCTIONS--------------------------------

def file_len(fname):
    with open(fname) as f:
        for i, l in enumerate(f):
            pass
    return i + 1

def grepWithLineNumbers(filename, mismatch, regex):
    regex = ['[{}]'.format(x) if len(x)>1 else x for x in regex]
    regex = ''.join(regex)
    process = subprocess.Popen(['agrep', '-n','-'+mismatch, regex, filename],
                               stdout=subprocess.PIPE, universal_newlines=True)
    stdout,stderr = process.communicate()
    lines = map(int,[x.split('\t')[0].split(':')[0] for x in stdout.split('\n')[:-1]])
    lines = []
    matches = 0
    for line in stdout.split('\n')[:-1]:
        splitLine = line.split('\t')[0].split(':')
        lines.append(int(splitLine[0]))
        try:
            matches += int(splitLine[1])
        except ValueError:
            matches+=1
    return matches, set(lines)

def possible_strings(individual):
    n = 0
    for group in individual:
        if len(group) == 1:
            if group == '.':
                n *= 20
            else:
                n += 1
        else:
            n *= len(group)
    return n

def separate_weighted_dict(dictionary):
    keysAndWeights = list(zip(*dictionary.items()))
    keys = keysAndWeights[0]
    weights = keysAndWeights[1]
    return keys, weights

# GA functions
# These make up the bulk of the actual GA algorithm.

def remove_letter(group,depth):
    """
    Helper function for mutation.
    Removes one letter from a group.
    If letter is the last one in group,
    returns empty string
    """
    if group == '.':

        letters, weights = separate_weighted_dict(letterProbabilities[depth])
        group = ''.join(np.random.choice(letters,5,replace=False,p=weights))
    elif len(group) == 1:
        return group
    else:
        group = list(group)
        del group[get_least_probable_index(group, depth)]
        group = ''.join(group)
    return group

def get_least_probable_index(group, depth):
    groupIndexProbabilities = {index: 1-letterProbabilities[depth][letter] for index, letter in enumerate(group)}
    probabilitySum = sum(groupIndexProbabilities.values())
    for index, probability in groupIndexProbabilities.items():
        groupIndexProbabilities[index] = probability/probabilitySum
    indexes, weights = separate_weighted_dict(groupIndexProbabilities)
    return np.random.choice(indexes, 1, p=weights)[0]

def get_letter_at_depth(depth):
    letters, weights = separate_weighted_dict(letterProbabilities[depth])
    return np.random.choice(letters,1,replace=False,p=weights)[0]

def change_letter(group,depth):
    group = list(group)
    n_letter = get_letter_at_depth(depth)
    while n_letter in group:
        n_letter = get_letter_at_depth(depth)
    group[get_least_probable_index(group, depth)] = n_letter
    group = ''.join(group)
    return group

def mutate(individual):
    """
    Function that performs mutation on individuals
    with the given mutation_chance probability
    Chooses randomly from three different mechanisms
    """
    m_individual = []
    mutation_chance = 1/len(individual)
    for depth in range(1, len(individual) + 1):
        c = individual[depth-1]
        if np.random.randint(0,100)/100 < mutation_chance:
            mut_choice = np.random.random()
            if mut_choice < 0.33: # Remove letter from group
                c = remove_letter(c,depth)
            elif 0.33 <= mut_choice < 0.66: # Add letter to group
                if c == '.':
                    m_individual.append(c)
                    continue
                n_letter = get_letter_at_depth(depth)
                c = ''.join(set(c).union(set(n_letter)))
            else:#CHANGING LETTER
                if c == '.':
                    c = remove_letter(c,depth)
                else:
                    c = change_letter(c, depth)
        m_individual.append(c)
    return m_individual

def crossover(individual1,individual2):
    """
    Crossover function.
    Switches random groups
    between two individuals
    """
    idx = np.random.choice(range(len(individual1)),0.3*len(individual1),replace=False)
    old_individual1 = individual1[:]
    try:
        for i in idx:
            individual1[i] = individual2[i]
            individual2[i] = old_individual1[i]
    except Exception as e:
        print(e)
        print('Error with individuals {} and {}'.format(individual1,individual2))
    return [individual1,individual2]

def crossover_at_index(individual1,individual2):
    idx = np.random.choice(range(len(individual1)-1),1,replace=False)
    old_individual1 = individual1[:]
    individual1[0:idx] = individual2[0:idx]
    individual2[0:idx] = old_individual1[0:idx]
    return [individual1,individual2]

candidates = pd.read_csv('data/specs_out.txt',sep='\t')['Pat?()'].values
    
```

```python
def get_edges(individual,vertex,i):
    if i == len(individual):
        return
    edges = []
    group = individual[i]
    if group == '.':
        edges.extend(list(vertex.out_edges()))
        for vertex in vertex.out_neighbours():
            d=get_edges(individual,vertex,i+1)
            if d != None:
                edges.extend(d)
    else:
        validEdges = []
        for letter in group:
            validEdges.extend([x for x in vertex.out_edges() if v_letter[x.target()] == letter])
        for edge in validEdges:
            d = get_edges(individual,edge.target(),i+1)
            if d != None:
                edges.extend(d)
        edges.extend(validEdges)
            
    return set(edges)


s1 = get_edges(individuals[3],root,0)
s2 = get_edges(individuals[4],root,0)

print(s1.intersection(s2))

def find_overlap(c1,c2):
    return len(get_edges(c1,root,0).intersection(get_edges(c2,root,0)))
```
